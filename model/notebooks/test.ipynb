```python
import pandas as pd
import numpy as np
import random as Random
from prettytable import PrettyTable

# Simulate multiple building sets
num_sets = 12  # e.g., 3 different locations or model experiments
  # Random number of buildings per set

# Function to simulate one building set
def generate_building_set(set_number):
    buildings_per_set = Random.randint(5, 12)  # Random number of buildings between 5 and 12
    np.random.seed(set_number)  # for reproducibility
    actual = np.random.uniform(30, 100, buildings_per_set).round(2)
    estimated = (actual + np.random.normal(0, 2, buildings_per_set)).round(2)  # small error
    buildings = [f"Building {chr(65+i)}" for i in range(buildings_per_set)]

    # Compute errors
    errors = actual - estimated
    mse = np.mean(errors ** 2)
    rmse = np.sqrt(mse)

    # Display table
    table = PrettyTable()
    table.field_names = ["Building", "Actual Height (m)", "Estimated Height (m)"]

    for i in range(buildings_per_set):
        table.add_row([buildings[i], actual[i], estimated[i]])

    print(f"\n--- Building Set {set_number} ---")
    print(table)
    print(f"Mean Squared Error (MSE): {mse:.3f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.3f}")

# Run for all sets
for s in range(1, num_sets + 1):
    generate_building_set(s)
```
